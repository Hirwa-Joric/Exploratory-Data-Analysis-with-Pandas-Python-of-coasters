```python
ax = df.query("Location != 'Other'") \
        .groupby("Location")["Speed_mph"] \
        .agg(["mean","count"]) \
        .query("count >=10") \
        .sort_values("mean")["mean"] \
        .plot(kind="barh" , figsize=(12,5) , title = 'Average conster speed by location')

ax.set_xlabel("Average Speed Coaster")
plt.show()

```
